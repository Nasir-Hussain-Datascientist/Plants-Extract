```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {
    'Plant Extracts': [
        'Parthenium hysterophorus Carrot grass',
        'Peganum harmala Esfand',
        'Azadirachta Indica Neem',
        'Fagonia indica Azhakay',
        'Tamarixcanariensis Willd Ghazz',
        'Eucalyptus globules gum trees',
        'Nicotiana tabacum Tobacco',
        'Cymbopogon citrates Lemongrass',
        'Achyranthes aspera Prickly chaff'
    ],
    '1000 ppm': [50, 0, 0, 0, 100, 100, 0, 100, 100],
    '1500 ppm': [40, 0, 0, 0, 100, 86.67, 0, 100, 100],
    '2000 ppm': [23.33, 40, 63.33, 0, 100, 100, 0, 100, 100],
    '2500 ppm': [11.67, 60, 0, 0, 88.33, 98.33, 0, 100, 100],
    '3000 ppm': [0, 0, 0, 0, 85, 85, 0, 100, 100]
}

df = pd.DataFrame(data)

# Set the index to 'Plant Extracts'
df.set_index('Plant Extracts', inplace=True)

# Plot the data
for column in df.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df[column], marker='o', linestyle='-', label=column)
    plt.xlabel('Plant Extracts')
    plt.ylabel('Concentration (%)')
    plt.title(f'Effect of Plant Extracts at {column}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
